```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import random
```

```python
SEED = 0
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```

```python
pdata = np.random.rand(6)
print(pdata)
pg = np.random.rand(6)
print(pg)
print(pdata/(pdata+pg))
```

```python
x = (np.arange(200)-100)/20
p = np.exp(-x**2)
q = np.exp(-(x-2)**2)/2
np.pi
```

```python
plt.plot(x, p)
plt.plot(x, q, ls=':')
plt.plot(x,p/(p+q))
plt.legend(['Pdata(x)','Pg(x)','Gd(x)'], loc='upper right')
plt.show()
```
